```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv ('data_1.csv',encoding='utf-8-sig')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('CA', axis=1, inplace=True)
df.drop('year', axis=1, inplace=True)
df.drop('month', axis=1, inplace=True)
df.drop('median_dom', axis=1, inplace=True)
```

```python
df.head()
```

```python
y = df['quality']
X = df.drop('quality',axis=1)
#, x_test, y_train, y_test = train_test_split(X, y, test_size=0.70, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.10, random_state=1)
```

```python
print("X_train: "+str(len(X_train)))
print("X_test: "+str(len(X_test)))
print("X_val: "+str(len(X_val)))
print("\n")
print("y_train: "+str(len(y_train)))
print("y_test: "+str(len(y_test)))
print("y_val: "+str(len(y_val)))
```

```python
logisticRegressionClf = LogisticRegression(solver='newton-cg',multi_class='multinomial')
```

```python
logisticRegressionClf.fit(X_train,y_train)

# Test Set
y_pred = logisticRegressionClf.predict(X_test)
print(classification_report(y_test, y_pred))


# Validation Set
y_val_pred = logisticRegressionClf.predict(X_val)
print(classification_report(y_val, y_val_pred))
```

```python
#Check train accuracy
y_train_pred = logisticRegressionClf.predict(X_train)
print(classification_report(y_train, y_train_pred))
```
